#### from number_items_traded -> score_traded

```python
TIER1_VALUE = 1
TIER2_VALUE = 3
TIER3_VALUE = 10

tier_items=pd.read_excel('exp.xlsx')
# print(tier_items)
tier3_items_set=set(list(tier_items["Tier 3"][tier_items["Tier 3"].notnull()].values))
tier2_items_set=set(list(tier_items["Tier 2"][tier_items["Tier 2"].notnull()].values))
tier1_items_set=set(list(tier_items["Tier 1"][tier_items["Tier 1"].notnull()].values))
print('tier3 items: ', tier3_items_set, len(tier3_items_set))
print('tier2 items: ', tier2_items_set, len(tier2_items_set))
print('tier1 items: ', tier1_items_set, len(tier1_items_set))
```

```python
def get_trade_value(row):
    offer_items_value = 0
    request_items_value = 0
    for ent in row['offer']:
        if ent['resource'] in tier1_items_set:
            offer_items_value += (ent['amount'] * TIER1_VALUE)
        elif ent['resource'] in tier2_items_set:
            offer_items_value += (ent['amount'] * TIER2_VALUE)
        else:
            offer_items_value += (ent['amount'] * TIER3_VALUE)
    for ent in row['request']:
        if ent['resource'] in tier1_items_set:
            request_items_value += (ent['amount'] * TIER1_VALUE)
        elif ent['resource'] in tier2_items_set:
            request_items_value += (ent['amount'] * TIER2_VALUE)
        else:
            request_items_value += (ent['amount'] * TIER3_VALUE)
    row['offer_value'] = offer_items_value
    row['request_value'] = request_items_value
    row['diff_value'] = np.abs(offer_items_value - request_items_value)
    return row
```

```python
accepted_trades_df = accepted_trades_df.apply(get_trade_value, axis=1)
print('avg. value differential = ', accepted_trades_df['diff_value'].mean())
print('std. value differential = ', accepted_trades_df['diff_value'].std())
print('median value differential = ', accepted_trades_df['diff_value'].median())
accepted_trades_df['diff_value'].hist()
```
